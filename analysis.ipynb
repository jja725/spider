<h4>Working with multiple stocks and OLS</h4>

```python
#Let's take a look at the solar sector
import datetime as dt
solar_df = data.DataReader(['FSLR', 'TAN','RGSE','SCTY'],'yahoo', start=dt.datetime(2013, 1, 1))['Adj Close']
#print(solar_df)
#Calculate returns
rets = solar_df.pct_change()
print(rets)
```

```python
rets['TAN']
```

```python
#Construct a scatter plot to compare returns on FSLR vs. returns on TAN
plt.scatter(rets.RGSE,rets.TAN)
```

```python
#Get correlations
solar_corr = rets.corr()
print(solar_corr)
```

```python
#Do some risk analysis
#We'll plot the mean and std or returns for each ticker to get a sense of the
#risk return profile
plt.scatter(rets.mean(), rets.std())
plt.xlabel('Expected returns')
plt.ylabel('Standard deviations')
for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (20, -20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
plt.show()
```

<h3>Simulating a random walk</h3>
<ul>
<li>Simulate a 1000 step equal probability 2-step random walk</li>
<li>Generate 50,000 walks</li>
<li>Compute the max and min of all walks (estimate of how far we can get)</li>
<li>Compute the average of the maximum and minimum of each walk</li>
<li>Compute the stdev of the max and min of each walk</li>

</ul>

```python
nwalks = 50000
nsteps = 1000

#Construct the walks
draws = np.random.randint(0,2,size=(nwalks,nsteps))
steps = np.where(draws>0,1,-1)
walks = steps.cumsum(1)

#Calculate the most walked in both directions
fartherest1 = walks.max()
fartherest2 = walks.min()
print("The most traveled:",fartherest1,fartherest2)

#Calculate the mean of the most walked for each walk in direction 1
indices=walks.argmax(axis=1) #Returns the indices of the max for each walk
max_mean = walks[np.arange(len(walks)) , indices].mean() #gets the mean of the values
max_std = walks[np.arange(len(walks)) , indices].std()
print(max_mean, max_std)

#Calculate the mean of the most walked for each walk in direction 2
indices = walks.argmin(axis=1)
min_mean = walks[np.arange(len(walks)) , indices].mean()
min_std = walks[np.arange(len(walks)) , indices].std()
print(min_mean,min_std)

```
